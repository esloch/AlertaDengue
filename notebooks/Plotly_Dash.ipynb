```python
# FAZ A BUSCA DE TODOS OS JSONS DISPONIVEIS
geojson_path = os.path.join(
    settings.STATIC_ROOT,
    'geojson',
    '*.json'
)

#geojson DICCIONARIO ONDE A CHAVE DO ITEM É O GEOCODE E O VALOR É O CONTEUDO DO ARQUIVO GEOJSON
geojson_files_rj = {
    'type': 'FeatureCollection',
    'features': []
}

for file_path in glob(geojson_path):
    try:    
        _geocode = file_path[-12:-5]
        int(_geocode)
    except:
        continue
        
    with open(file_path, 'r') as f:
        content = f.read()
                
        geojson = json.loads(content)
        # create the id attribute
        geojson['features'][0]['id'] = (
            str(geojson['features'][0]['properties']['geocodigo'])
        )
        
        # if geocode == _geocode:
        #     with open('/tmp/geocode-{}'.format(geocode), 'w') as fw:
        #         json.dump(geojson, fw, indent=2)

        geojson_files_rj['features'].append(geojson['features'][0])

        
# geojson_files_rj[geocode]    
```

```python
from urllib.request import urlopen
import json

with open('/tmp/geojson-counties-fips.json', 'w') as f:
    with urlopen(
        'https://raw.githubusercontent.com/plotly/datasets/'
        'master/geojson-counties-fips.json'
    ) as response:
        counties = json.load(response)
        json.dump(counties, f, indent=2)
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/fips-unemp-16.csv",
                   dtype={"fips": str})

import plotly.graph_objects as go
```

```python
df_rj.loc[3300100020180518058, 'level_alert'] = 2

fig = go.Figure(go.Choroplethmapbox(
    geojson=geojson_files_rj, 
    locations=df_rj.municipio_geocodigo, 
    z=df_rj.level_alert,
    colorscale="Viridis", zmin=0, zmax=12,
    marker_opacity=0.5, marker_line_width=1))

fig.update_layout(mapbox_style="carto-positron",
                  mapbox_zoom=7, mapbox_center = {"lat": -22.187, "lon": -43.176})
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
